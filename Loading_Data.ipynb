```python
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
breast_cancer_wisconsin_diagnostic = fetch_ucirepo(id=17) 
  
# data (as pandas dataframes) 
X = breast_cancer_wisconsin_diagnostic.data.features 
y = breast_cancer_wisconsin_diagnostic.data.targets 
```

```python
import matplotlib.pyplot as plt

count_per_diagnosis = y.groupby('Diagnosis').size()
diagnosis = y.groupby('Diagnosis').size().index.values

plt.bar(diagnosis, count_per_diagnosis)
plt.show()
```

```python
fig, ax = plt.subplots(15,2, figsize=(10,20)) 

i = 0
j = 0
for col in X.columns:
    ax[i,j].hist(X[col], bins=30)
    ax[i,j].grid()
    ax[i,j].set_title(col)
    j += 1
    if j == 2:
        i += 1
        j = 0

fig.tight_layout(pad=1.0)
plt.show()
```

```python
fig, ax = plt.subplots(15,2, figsize=(10,50)) 

i = 0
j = 0
for col in X.columns:
    ax[i,j].boxplot(X[col])
    ax[i,j].grid()
    ax[i,j].set_title(col)
    j += 1
    if j == 2:
        i += 1
        j = 0

fig.tight_layout(pad=1.0)
plt.show()
```
